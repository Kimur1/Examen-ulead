```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from scipy.optimize import fsolve

# Datos
years = np.array([1910, 1930, 1950, 1970, 1980, 1990, 2000])
recaudaciones = np.array([0.7, 4.1, 39.4, 192.8, 517.1, 1032.0, 2025.0])
x = np.array([0, 20, 40, 60, 70, 80, 90])  # Años desde 1910

# Funciones
def f(x):
    return 0.786 * (1.094 ** x)

def g(x):
    return 0.503 * x**3 - 27.3 * x + 149.2

# Valores de x para las funciones
x_vals = np.linspace(0, 90, 1000)

# Crear el DataFrame
df = pd.DataFrame({
    'Año': years,
    'Recaudaciones': recaudaciones
})

# Calcular los valores de las funciones
f_vals = f(x_vals)
g_vals = g(x_vals)

# Graficar los datos y las funciones
fig = go.Figure()

# Datos reales
fig.add_trace(go.Scatter(x=years, y=recaudaciones, mode='markers', name='Datos Reales'))

# Función f(x)
fig.add_trace(go.Scatter(x=1910 + x_vals, y=f_vals, mode='lines', name='f(x) = 0.786(1.094)^x'))

# Función g(x)
fig.add_trace(go.Scatter(x=1910 + x_vals, y=g_vals, mode='lines', name='g(x) = 0.503x^3 - 27.3x + 149.2'))

# Encontrar el año en que se recaudó $1 billón
f_billion_year = 1910 + fsolve(lambda x: f(x) - 1000, 50)[0]
g_billion_year = 1910 + fsolve(lambda x: g(x) - 1000, 50)[0]

fig.add_trace(go.Scatter(x=[f_billion_year], y=[1000], mode='markers+text', text=['f(x) = 1000'], name='f(x) = 1000', textposition="bottom center"))
fig.add_trace(go.Scatter(x=[g_billion_year], y=[1000], mode='markers+text', text=['g(x) = 1000'], name='g(x) = 1000', textposition="bottom center"))

# Ajustes del gráfico
fig.update_layout(
    title='Recaudaciones del Gobierno Federal',
    xaxis_title='Año',
    yaxis_title='Recaudaciones (miles de millones de dólares)',
    legend_title='Leyenda'
)

# Mostrar el gráfico
fig.show()

# Determinar cuál de las dos funciones modela mejor los datos
from sklearn.metrics import mean_squared_error

# Calcular los errores cuadráticos medios
mse_f = mean_squared_error(recaudaciones, f(x))
mse_g = mean_squared_error(recaudaciones, g(x))

# Mostrar cuál función modela mejor los datos
mejor_funcion = 'f(x)' if mse_f < mse_g else 'g(x)'
print(f"La función que mejor modela los datos es {mejor_funcion} con un MSE de {min(mse_f, mse_g)}")
```
